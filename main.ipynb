```python
apartaments = data.copy()
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
inverse_mappings = {}

for col in data.columns:
    if data[col].dtype == 'object':
        data[col] = le.fit_transform(data[col])
        inverse_mappings[col] = {v: k for k, v in enumerate(le.classes_)}
```

```python
# model de regresie liniara
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

data.head()
X = data.drop(['SalePrice'], axis=1)
X.head()
```

```python
y = data['SalePrice']
y.head()
```

```python
test_data = pd.read_csv('Data/test.csv')
test_data.head()
```

```python
test_data = test_data.drop(['Id'], axis=1)
```

```python
test_data['LotFrontage'] = test_data['LotFrontage'].fillna(test_data['LotFrontage'].mean())
```

```python
model = LinearRegression()
X = X.fillna(0)
model.fit(X, y)
```

```python
test_data = test_data.fillna(0)
# inlocuim obiectele cu valori numerice conform inverse_mappings
for col in test_data.columns:
    if test_data[col].dtype == 'object':
        test_data[col] = test_data[col].map(inverse_mappings[col])
predictions = model.predict(test_data)
predictions
```

```python
predictions.shape
```

```python
# punem la predictii o coloana cu id-urile, pornind de la 1461
predictions = pd.DataFrame(predictions, columns=['SalePrice'])
predictions['Id'] = range(1461, 2920)
predictions = predictions[['Id', 'SalePrice']]
predictions
```

```python
predictions.to_csv('Data/predictions1.csv', index=False)
```

```python
# folosim regresia Ridge
from sklearn.linear_model import Ridge
model = Ridge()
model.fit(X, y)
predictions = model.predict(test_data)
predictions = pd.DataFrame(predictions, columns=['SalePrice'])
predictions['Id'] = range(1461, 2920)
predictions = predictions[['Id', 'SalePrice']]
predictions.to_csv('Data/predictions2.csv', index=False)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X, y)
predictions = model.predict(test_data)
predictions = pd.DataFrame(predictions, columns=['SalePrice'])
predictions['Id'] = range(1461, 2920)
predictions = predictions[['Id', 'SalePrice']]
predictions.to_csv('Data/predictions3.csv', index=False)
```
